```python
# Libraries 

import pandas as pd
```

```python
# -- Load Data --
file_path = '../reports/PharmaReach_Final_Strategy_Dashboard.csv'
df = pd.read_csv(file_path)
```

```python
# -- data overview --
print(df.info())
```

```python
# Function get top targets by indication

def get_top_targets(city_name, specialty_name=None):
    """
    Returns the top 5 physicians in a specific city, 
    optionally filtered by specialty.
    """
    results = df[df['city'] == city_name.title()]
    
    if specialty_name:
        results = results[results['specialty'] == specialty_name.title()]
        
    return results.sort_values(by='lead_score', ascending=False).head(5)

```

```python
display(get_top_targets('New York', 'Cardiology'))
```

```python
display(get_top_targets('New York', 'Oncology'))
```
